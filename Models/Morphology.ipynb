```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib notebook
from skimage.io import imread,imshow,imsave
```

```python
im = imread(r'C:\Users\LENOVO\Desktop\perio.png')
bw = im>127

plt.figure()
plt.gray()
plt.imshow(bw)
plt.show()
```

```python
from skimage.morphology import disk

se = disk(5)
plt.figure()
plt.imshow(se)
plt.show()
```

```python
from skimage.morphology import opening,closing,erosion,dilation

plt.figure()
plt.subplot(2,2,1)
plt.imshow(erosion(bw, se))
plt.subplot(2,2,2)
plt.imshow(dilation(bw, se))
plt.subplot(2,2,3)
plt.imshow(opening(bw, se))
plt.subplot(2,2,4)
plt.imshow(closing(bw, se))
plt.show()
```

```python
clean = closing(opening(bw,se),se)
plt.figure()
plt.subplot(1,2,1)
plt.imshow(bw)
plt.subplot(1,2,2)
plt.imshow(clean)
plt.show()
```
